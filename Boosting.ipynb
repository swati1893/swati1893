Boosting:
1. Boosting is sequential process of model creation
2. Boosting follows emsembled algorithm
3. It create mulitple sample and small decision tree
4. Each sample created based on error generated from previous sample
5. Each decision tree tries to reduce error availbale in previous DT

Types of Boosting:
1. Adaptive Boosting (Ada)--it will consist of small decsion tree with one root node and two leaf node, is called Decision stump(not fully grown)
2. Gradient Boosting
3. Extreme Gradient boosting (XG)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

import warnings
warnings.filterwarnings("ignore")
```

```python
#Removing outliers
#Removing skewness
```

```python
#feature and target selection
X=df.iloc[:,:-1] #feature
y=df.iloc[:,-1]  #target
```

```python
#train test split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,train_size=0.3,random_state=1)
```

Ada boost classifier model

```python

#n_estimators:decision stump-->50 by default
#depth=1
model=AdaBoostClassifier(n_estimators=100,random_state=1)
```

```python
#training model
model.fit(xtrain,ytrain)
```

```python
#prediction
ypred=model.predict(xtest)
print(classification_report(ytest,ypred))
```
