Handling missing values--
If values are not entered in the field, such value are called as missing value
We can fill such value by NAN(not a number)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from sklearn.impute import SimpleImputer
```

```python
#dropna--drop missing value
df_csv.dropna()
```

#Rules to be followed while dropping missing values
1.If missing value is less than 3% then use dropna
2.If missing value is 4-50% then use fillna()--(recommended)---syntax fillna(value to be replaced)
3.If missing value is greater than 50% then drop entire column--(not recommended)

```python
pmean=df_csv['Python'].mean()
pmean

mmean=df_csv['Machine Learning'].mean()
mmean

#df_csv['Python'].value_counts()
#df_csv['Python'].info()
#df_csv['Python'].describe()
#print(df_csv.isna().sum())

df_csv['Python']=df_csv['Python'].fillna(pmean)

df_csv['Machine Learning']=df_csv['Machine Learning'].fillna(mmean)

df_csv


```

```python
#fillna()--fills missing value with new values

#we are calculating  mean for each NAN
pymean=df_csv['Python'].mean()
pymean
mlmean=df_csv['Machine Learning'].mean()
mlmean

#we are replacing NAN by mean
df_csv['Python']=df_csv['Python'].fillna(pymean)
df_csv['Machine Learning']=df_csv['Machine Learning'].fillna(mlmean)

# check result
df_csv
```

Skewness--
1. Measure of Assymmetry
2. It only applicable on numerical column
3. Mean>Median--Positive sknewnes
4. Mean<Median--Negative skewness
5. Mean=Mode=Median---No sknewness
6. We used two method to remove it
    1. log value
    2. Squareroot value

```python
#importing scipy library and included only numerical column
from scipy.stats import skew
numcol=feature.select_dtypes(include=['int64','float64']).columns
numcol
```

```python
#Accessing all columns from numcol array
for col in feature[numcol]:
    print(col)
    print(skew(feature[col]))
    plt.figure()
    sns.distplot(feature[col])
    plt.show()
```

Rules for removing skewness----
1. If the column has the skewness shoing good correlation with target, don't remove skewness
2. If the column has the skewness has negative value, don't remove skewness
3. If all the columns are highly skewed, we are unable to remove skewness for each column, so check correlation between feature and target values

```python
pd.concat([feature[numcol],target],axis=1).corr()
```

```python
pd.concat([feature[numcol],target],axis=1).corr()['price'].sort_values()
```

```python
#removing the skewness using logvalue method
feature['symboling']=np.log(feature['symboling'])
```

1.One-hot/dummy encoding
2.Label/ordinal encoding
3.Target encoding
4.Frequency/count encoding
5.Binary encoding
6.Feature encoding

Data scaling/Feature scaling
It is a process of bringging all the numeric quantity under one standard scale
It can be performed during pre processing to handle highly varying units
If data scaling is not done then ML model assumes greater number as large and smaller number as small irrespective of the units.
Data scaling is not required in non-linear model.It is required in linear model(linear regression).
There are 2 types of data scaling
Min-max scaler
standard scaler

```python
n1=10   # kg
n2=1000 # gram
n1>n2    #need data scalling
```

```python
a=np.array([1,2,3,4,5])
b=np.array([10,20,30,40,50])
c=np.array([100,200,300,400,500])

def minmax(arr):
    xnew=(arr-arr.min())/(arr.max()-arr.min())
    print(xnew)
```

```python
minmax(a)
```

```python
minmax(b)
```

```python
minmax(c)
```
