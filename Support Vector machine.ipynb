SVM:
1. SVMs can be used for a variety of tasks, such as text classification, image classification, spam detection, handwriting identification, gene     expression analysis, face detection, and anomaly detection
2. It's main objective is to find optimal boundry which seperates datapoints
3. Boundry also called Hyperplane
4. 2D data have line
5. 3D data have hyperplane
6. In Safe boundry datapoints has good margin from line or hyperplane
7. SVM can create multiple boundries
8. Data points near to boundry are called as support vector
9. If data points are non linearly seperable, we can use kernel function
     1. Polynomial kernel function
     2. Radial basis kernel function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
```

```python
#reading the dataset
df=pd.read_csv(r"H:\My Drive\ML\ML datasources\bank.csv")
df
```

```python
#feature and target selection
X=df.iloc[:,:-1] #feature
y=df.iloc[:,-1]  #target
y.value_counts()
```

```python
#train test split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Model building and training
lin_svc=LinearSVC(random_state=1)
lin_svc.fit(xtrain,ytrain)
```

```python
#testing
ypred=lin_svc.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
#C: is a parameter which has to be added to
#improve regularization(make margin as soft margin)
#C value low:good regularization
#C value=1 indicates no regularization
```

```python
#soft margin(improving regularization)
#linear svc model with soft margin
lin_svc_soft=LinearSVC(random_state=1,C=0.8)
lin_svc_soft.fit(xtrain,ytrain) #training model
```

```python
#testing model
ypred=lin_svc_soft.predict(xtest)
print(classification_report(ytest,ypred))
#as the accuracy is decreasing we can conclude that
#the model is not overfitting
```

```python
#polynomial kernel function
svc=SVC(kernel='poly',degree=2,random_state=1)
#polynomial-n dimensional 4x**3+2x**2+3x+2-->3
svc.fit(xtrain,ytrain)
```

```python
#testing the model
ypred=svc.predict(xtest)
print(classification_report(ytest,ypred))
```

```python
#radial basis kernel function
svc1=SVC(kernel='rbf',random_state=1) #radial basis model
svc1.fit(xtrain,ytrain)
```

```python
#testing model
ypred=svc1.predict(xtest)
print(classification_report(ytest,ypred))
```
